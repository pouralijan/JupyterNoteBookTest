```python
import requests
import bs4
import re

not_found_count = 0

url = 'https://git.ir/python/page/{}/'
for i in range(1,100):
    new_url = url.format(i)
    print(new_url)
    page = requests.get(new_url)
    if page.status_code == 404:
        print(new_url, '>>> Not found, 404' )
        not_found_count +=1
        print(not_found_count)
        if not_found_count > 4:
            break
        continue
    soup = bs4.BeautifulSoup(page.content, "html.parser")
    objs += soup.select("#posts-container a")[:-1]
    #print(objs)




#print(names)
print(len(objs))
for obj in objs:
    if obj.text == '':
        continue
    print ('{} -> {}'.format(obj.text, obj.get('href'),))
objs = []
```
